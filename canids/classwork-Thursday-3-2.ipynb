```python
canid_file = open("no_headers_canid_data.csv", "r")
canid_list = canid_file.readlines()

from collections import defaultdict

species_ranges = defaultdict(list)
for line in canid_list:
    species = line.split(",")[5]
    max_year = line.split('","')[10]
    min_year = line.split('","')[11]
    mean_age = (float(max_year) + float(min_year))/2
    
    species_ranges[species].append(mean_age) # add species as keys and mean ages as elements to the list of ages
```

```python
species_ranges.values()
```

```python
# get the oldest and youngest ages of fossils for each species
# in species_ranges
output = open("canid_ranges.csv", "w")
for key in species_ranges.keys():
    ages = species_ranges[key]
    min_age = min(ages)
    max_age = max(ages)
    species = key
    genus = key.split(" ")[0]
    outline = "{},{},{},{}\n".format(genus, species, min_age, max_age)
    output.write(outline)
```
