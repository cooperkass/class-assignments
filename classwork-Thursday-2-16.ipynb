```python
def makeBirdDict(filename):
    # this function opens the data file from Thomas et al and
    # extracts the species and family as a dictionary
    birdDict = {}
    # open the file
    fobj = open(filename, "r", encoding = "ISO-8859.15")
    # read the file and create a list of the file
    store_list = fobj.readlines()[2:]
    # split lines on comma to make new lists
    for line in store_list:
        species = line.split(",")[0]
        family = line. split(",")[2]
        # stick species and family into birdDict
        birdDict[species] = family
    return birdDict
```

```python
thomas_birds = makeBirdDict("formatted.nature.csv")
```
